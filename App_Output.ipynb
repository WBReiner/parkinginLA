```python
df = user_SUBSET
df= df[['hour', 'violationdescription', 'result', 'weekdayindex']]
df.columns = ['Hour', 'Description', 'Result', 'Day']
```

```python
# Make a new column that "bins" the results into risk categories. 
# The process uses the np.digitize function:
# https://docs.scipy.org/doc/numpy/reference/generated/numpy.digitize.html
# Pass in the numeric data frame as a "float" dtype and then pass in an np.array with the bin edges
# For the specifics of the convention, check the above link. But anything below the first number will 
# be in the 'zero' bin, anything above (or equal to) that value but below the next will be in the '1' bin and so on.

df['bins'] = np.digitize(np.asarray(df['Result'], dtype = 'float'), np.asarray([1, 1.1, 5.1,10,26,51,76,]))
df.head()
```

```python
df['daybins'] = np.digitize(np.asarray(df['Day'], dtype = int), np.asarray([1,2,3,4,5,6,7,]))
df.head()
```

```python
# Now, create a dictionary with the keys as the bin numbers and values as the risk category
# Then map the bins with that diction into a new df column

dict_days = {0: "Monday", 1: "Tuesday", 2: "Wednesday", 3: "Thursday", 4:"Friday", 5:"Saturday", 6:"Sunday"}
df['Day'] = df['daybins'].map(dict_days)
df.head()
```
