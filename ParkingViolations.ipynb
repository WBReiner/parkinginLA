For your dataset, the easiest way to get the time into datetime format would be to convert it to a string and use strptime to pull hours and minutes. Then datetime will keep track of everything in one easy-to-use format. You can even combine date info into the same object

Not sure how your dates are formatted, but it will probably also be easier to include them in the same string so you only have to do a single strptime

but, just as an example, for your times, call your initial value time_float, you'd do:

'import datetime as dt

time_str = str(time_float)
time_obj = dt.datetime.strptime(time_str, '%H%M')'

(edited)
that second argument is the format of your time string

datetime_str = date_str + '  ' + time_str   #combines the two strings with a space in the middle for easier reading
